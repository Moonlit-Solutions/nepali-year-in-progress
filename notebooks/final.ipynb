```python
import datetime as dt
import nepali_datetime
```

```python
def convert(val):
    val = str(val)
    mapping = {
    "0":"०",
    "1":"१",
    "2":"२",
    "3":"३",
    "4":"४",
    "5":"५",
    "6":"६",
    "7":"७",
    "8":"८",
    "9":"९"
    }
    return ''.join([mapping[char] for char in val])
```

```python
comp = ['█','░']

threshold_dict = {}
for i in range (1, 366):
    day_of_year = i
    number_of_days = 365
    shrink_factor = 4
    shrink = 100/shrink_factor
    threshold_dict[str(round((day_of_year/number_of_days)*100))] = [day_of_year, (comp[0]*round((day_of_year/number_of_days)*shrink)+comp[1]*round((1-(day_of_year/number_of_days))*shrink)+' '+convert(round((day_of_year/number_of_days)*100))+'%')]

    
final_dict = {}
for key in threshold_dict:
    final_dict[threshold_dict[key][0]] = threshold_dict[key][1]
```

```python
final_dict
```

```python
# day_of_year = 286
day_of_year = (nepali_datetime.date.today() - nepali_datetime.date(nepali_datetime.date.today().year, 1, 1)).days 
day_np = convert(nepali_datetime.date.today().day)
month_np = convert(nepali_datetime.date.today().month)
year_np = convert(nepali_datetime.date.today().year)
```

```python
date_np = day_np+'-'+month_np+'-'+year_np
```

```python
if day_of_year in final_dict:
    print ('आज: '+date_np+"\n"+final_dict[day_of_year])
```

```python
if day_of_year in final_dict:
    tweet_text = """
आज: {}
{}
    """.format(date_np, final_dict[day_of_year])
```
